```python
import pandas as pd
import numpy as np
```

```python
# Đọc dữ liệu
df = pd.read_csv('project_heart_disease.csv')
display(df.info())
display(df.head())

#Xử lí giá trị thiếu
missing_values = df.isnull().sum()
missing_values_count = missing_values.sum()
print("Missing Values:\n", missing_values)
print(f"Total missing values: {missing_values_count}")
for col in df.select_dtypes(include=[np.number]).columns:
    df[col].fillna(df[col].mean(), inplace = True)
for col in df.select_dtypes(include=['object']).columns:
    df[col].fillna(df[col].mode()[0], inplace=True) 

#Xử lí giá trị trùng lặp
duplicate_count = df.duplicated().sum()
print(f"Total duplicate values:{duplicate_count}")
df = df.drop_duplicates()

#Xuất ra file mới
df.to_csv('Preprocessing_project_heart_disease.csv', index = False)

# Kiểm tra lại giá trị thiếu sau khi xử lý
missing_values_after = df.isnull().sum()
missing_values_count_after = missing_values_after.sum()
print(f"Total missing values after processing: {missing_values_count_after}")

# Kiểm tra lại giá trị trùng lặp sau khi xử lý
duplicate_count_after = df.duplicated().sum()
print(f"Total duplicate values after processing: {duplicate_count_after}")
```

```python
df = pd.read_csv('Preprocessing_project_heart_disease.csv')
summary = pd.DataFrame({
    "Field Name": df.columns,
    "NULL": df.isnull().sum().values,
    "Missing": df.isnull().sum().values,
    "Actual": len(df) - df.isnull().sum().values,
    "Completeness": ((len(df) - df.isnull().sum()) / len(df) * 100).round(2).astype(str) + "%",
    "Cardinality": df.nunique().values,
    "Uniqueness": ((df.nunique() / len(df)) * 100).round(2).astype(str) + "%",
    "Distinctness": ((df.nunique() / len(df)) * 100).round(2).astype(str) + "%"
})
display(summary)
```
